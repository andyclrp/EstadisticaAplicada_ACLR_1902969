Vamos a crear un dataframe
luego lo graficamos

```python
mun_nl= pd.DataFrame({'municipio':['Mty','Gpe', 'Apo','Snl','Sant'], 'nacimientos':[500,658,445,345,454]})
mun_nl
```

```python
#grafico simple
mun_nl = mun_nl.sort_values('nacimientos', ascending=False)
plt.bar(mun_nl['municipio'],
        mun_nl['nacimientos'],
        color=['green','black','black','black','black'])

#agregar titulo
plt.title('Nacimientos x municipio')

#agregar nombre al eje x
plt.xlabel('municipio')

#agregar nombre al eje y
plt.ylabel('cant')

#cambiar saltos escala y
plt.yticks([0,200,400,600,800])

plt.show()
```

```python
#fig es el nombre del lienzo
fig, (panel1,panel2)=plt.subplots(2,1) #subgraficas, su nombre, filas,columnas
fig.tight_layout ()

#panel 1 (subgrafica 2)
x1=x
y1=y
panel1.plot(x,y,label='nombre de linea' )
panel1.title.set_text('Grafica 1')

#panel 2 (subgrafica 2)
x2 = mun_nl ['municipio']
y2 = mun_nl ['nacimientos']
panel2.bar(x2,y2)
panel2.title.set_text('Grafica 2')
```

```python
mun_nl = mun_nl.sort_values('nacimientos', ascending=True)
plt.barh(mun_nl['municipio'],
        mun_nl['nacimientos'],
        color=['green','black','black','black','black'])
plt.xticks([0,200,400,600,800])
```

```python
mun_nl = mun_nl.sort_values('nacimientos', ascending=True)
plt.scatter(mun_nl['municipio'],
        mun_nl['nacimientos'],
        color=['green','black','blue','yellow','brown'])
plt.yticks([0,200,400,600,800])
```
