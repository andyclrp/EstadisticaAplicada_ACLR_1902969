```python
# Importamos librerias 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Librerias de supervivencia
from lifelines import KaplanMeierFitter
from lifelines.statistics import logrank_test
from lifelines import CoxPHFitter
```

```python
#importamos un archivo de base de datos-> km_relacion.csv
#datos
relacion = pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/km_relacion.csv')
```

```python
#imprimos la base de datos
relacion.head()
```

La base de datos consiste en 42 observaciones.<br> Es de un experimento, donde se dividió a las observaciones en dos grupos, cada uno de 21 observaciones.
****
**tratamiento** : 1 les tocó ser detallistas (dar regalo en transcurso de la semana),  0 no son detallistas (ningún regalo) <br>
**t** : Tiempo de la relación en semanas <br> 
**estado** : 1 si se termino la relación, 0 datos censurados o no sabemos si se terminó
****
**Hipótesis** :
La relación dura más si las personas son detallistas con la pareja

```python
#Estadísticas descriptivas
relacion.describe()
```

```python
relacion.groupby('tratamiento').agg(['mean','std','count'])
```

```python
detallistas = relacion.query ('tratamiento==1')
no_detallistas = relacion.query ('tratamiento==0')
```

```python
detallistas
```

```python
no_detallistas
```

```python
km_d = KaplanMeierFitter().fit(detallistas['t'], detallistas['estado'])
km_nd = KaplanMeierFitter().fit(no_detallistas['t'], no_detallistas['estado'])
```

```python
#Graficamos las funciones de Supervivencia
km_d.plot_survival_function(label = 'detallistas')
km_nd.plot_survival_function(label = 'no detallistas')
plt.show()
```

```python
lrt = logrank_test(detallistas['t'], detallistas['estado'],
                   no_detallistas['t'], no_detallistas['estado'])
lrt.print_summary()
```

Ho: funciones de supervivencia = <br>
Ha: funciones de supervivencia $\neq$

A un nivel de confianza del 95% se rechaza Ho, por lo tanto las funciones de supervivencia de detallistas y no detallistas son distintas

```python
cph = CoxPHFitter().fit(relacion, 't', 'estado')
cph.print_summary()
```

1 - exp(coef) = 1 - 0.21 = 0.79 <br>
Ser detallista disminuye el riesgo de que la relacion termine. <br>
En promedio alguien que es detallista tiene 79% menos de riesgo de que termine la relacion respecto a alguien que no es detallista.
