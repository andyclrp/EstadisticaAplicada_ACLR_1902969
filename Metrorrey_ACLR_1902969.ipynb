```python
# Importamos librerias 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#librerias especializadas
from statsmodels.tsa.api import seasonal_decompose
```

```python
#datos
metrorrey = pd.read_csv("metrokk.csv")
```

```python
metrorrey.head()
```

Siempre que trabajemos con series de tiempo debemos declarar una variable con formato de tiempo y debemos convertirla en el index

```python
#en fecha con pandas
metrorrey['Mes'] = pd.to_datetime(metrorrey['Mes'], format='%Y/%m')
```

```python
metrorrey = metrorrey.set_index('Mes')	
```

```python
metrorrey.index
```

Si el dataframe ya tiene una variable de fecha, una frecuencia y es el index; ya se pueden emplear las herramientas de análisis de series de tiempo

```python
#graficamos
metrorrey['Pasajeros'].plot()
plt.title('Miles de pasajeros en metrorrey\n 2008-2023')
plt.ylabel('Miles')
plt.show()
```

```python
metrorrey['Ingresos'].plot()
plt.title("Ingresos mensuales de metrorrey\n20008-2023")
plt.ylabel('Miles de pesos')
plt.show()
```

```python
series1= seasonal_decompose(metrorrey['Pasajeros'], model='ad')
series1.plot()
plt.show()
```

```python
#revisar en  que mes hay mas pasajeros en metrorrey
series1.seasonal.iloc[0:12]
```

Meses con menos pasajeros: marzo, mayo, agosto, octubre, noviembre

Meses con más pasajeros: enero y diciembre
